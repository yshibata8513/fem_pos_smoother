```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_curvature_2_derivative(x, y):

    curvatures = [0.0]
    for i in np.arange(1, len(x)-1):
        dxn = x[i] - x[i - 1]
        dxp = x[i + 1] - x[i]
        dyn = y[i] - y[i - 1]
        dyp = y[i + 1] - y[i]
        dn = np.hypot(dxn, dyn)
        dp = np.hypot(dxp, dyp)
        dx = 1.0 / (dn + dp) * (dp / dn * dxn + dn / dp * dxp)
        ddx = 2.0 / (dn + dp) * (dxp / dp - dxn / dn)
        dy = 1.0 / (dn + dp) * (dp / dn * dyn + dn / dp * dyp)
        ddy = 2.0 / (dn + dp) * (dyp / dp - dyn / dn)
        curvature = (ddy * dx - ddx * dy) / ((dx ** 2 + dy ** 2) ** 1.5)
        curvatures.append(curvature)
    return curvatures
```

```python
dtheta = 0.01
radius = 50
N = 200
theta = torch.arange(N)*dtheta
theta[::2] += 0.4*dtheta
noise_x = torch.randn(N)*0.01
noise_y = torch.randn(N)*0.01
x = torch.cos(theta)*radius + noise_x
y = torch.sin(theta)*radius + noise_y
c = calc_curvature_2_derivative(x,y)
#plt.plot(x,y)
plt.plot(c)
```

```python
dx = x[1:] - x[:-1]
dy = y[1:] - y[:-1]
dist = torch.sqrt(dx*dx + dy*dy)
```

```python
N = len(x)
weights = torch.zeros(N -2)
for i in range(N-2):
    weights[i] = dist[i]/(dist[i]+dist[i+1])
```

```python
sigma = 10.1
A = torch.eye(N,N)*1.0/sigma
```

```python
def Set0(A):
    i=0
    w_pm = weights[i]
    w_pp= 1.0 - weights[i]

    A[i,i]   = 1.0/sigma -w_pm*w_pm + 1.0
    A[i,i+1] = w_pm
    A[i,i+2] = -w_pp*w_pm
    return A
```

```python
def Set1(A):
    i=1
    w_m  = weights[i-1]
    w_p  = 1.0 - weights[i-1]
    w_pm = weights[i]
    w_pp= 1.0 - weights[i]

    A[i,i-1] = - w_m
    A[i,i]   = 1.0/sigma -w_pm*w_pm + 1.0
    A[i,i+1] = w_pm - w_p
    A[i,i+2] = -w_pp*w_pm
    return A
```

```python
def Set_N2(A):
    i=N-2
    w_mm = weights[i-2] 
    w_mp = 1.0 - weights[i-2]
    w_m  = weights[i-1]
    w_p  = 1.0 - weights[i-1]

    A[i,i-2] = -w_mm*w_mp 
    A[i,i-1] = w_mp - w_m
    A[i,i]   = 1.0/sigma -w_mp*w_mp + 1.0
    A[i,i+1] =- w_p
    return A
```

```python
def Set_N1(A):
    i=N-1
    w_mm = weights[i-2] 
    w_mp = 1.0 - weights[i-2]

    A[i,i-2] = -w_mm*w_mp 
    A[i,i-1] = w_mp
    A[i,i]   = 1.0/sigma -w_mp*w_mp+ 1.0
    return A
```

```python
for i in range(2,N-2):
    w_mm = weights[i-2] 
    w_mp = 1.0 - weights[i-2]
    w_m  = weights[i-1]
    w_p  = 1.0 - weights[i-1]
    w_pm = weights[i]
    w_pp= 1.0 - weights[i]
    A[i,i-2] = -w_mm*w_mp 
    A[i,i-1] = w_mp - w_m
    A[i,i]   = 1.0/sigma -w_mp*w_mp -w_pm*w_pm + 1.0
    A[i,i+1] = w_pm - w_p
    A[i,i+2] = -w_pp*w_pm
```

```python
bx = 1.0/sigma*x.view(-1,1)
by = 1.0/sigma*y.view(-1,1)
B  = torch.cat([bx,by],dim=-1)
```

```python
X, LU = torch.solve(B, A)
```
